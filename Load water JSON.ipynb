```python
import numpy as np

from ipywidgets import (interact)

import bokeh
from bokeh.plotting import (figure, show, output_notebook)
from bokeh.models import (LinearColorMapper, Ticker, ColorBar)
from bokeh.io import (push_notebook)

output_notebook()

import datetime
import struct
import base64
```

```python
import json
```

```python
data = json.load(open('data.json'))['data']
```

```python
def parse_time(s):
    return datetime.datetime.strptime(s, '%Y-%m-%dT%H:%M:%S.%fZ').timestamp()
```

```python
def parse_data(b):
    #return b 
    return struct.unpack('>II', base64.b64decode(b))
```

```python
q = np.array([(parse_time(x['timestamp']),) + parse_data(x['data_raw']) for x in data], dtype=int)
```

```python
p = figure(title="data", x_axis_label='x', y_axis_label='y')
# p.line(q[:,0], q[:,1], legend="q1", line_width=2)
p.line(q[:,1], q[:,2], legend="q2", line_width=2)
show(p)
```
